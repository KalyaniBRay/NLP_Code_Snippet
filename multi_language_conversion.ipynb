# Translate English to Hindi

```python
from transformers import MarianMTModel, MarianTokenizer

# Load pre-trained model and tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-hi'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_text(text, model, tokenizer):
    # Tokenize and translate
    tokens = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text

# Example usage
english_text = "Hello, how are you?"
hindi_translation = translate_text(english_text, model, tokenizer)
print("Hindi Translation:", hindi_translation)
```

# Line by Line translate English to Hindi

```python
from transformers import MarianMTModel, MarianTokenizer

# Load pre-trained model and tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-hi'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_text(text, model, tokenizer):
    # Tokenize and translate
    tokens = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text


def translate_file_line_by_line(file_path, model, tokenizer):
    with open(file_path, 'r', encoding='utf-8') as ipfile:
        for line in ipfile:
            line = line.strip()
            with open(output_file, 'a') as opfile:
                if line:
                    translated_line = translate_text(line, model, tokenizer)
                    print(f"Original: {line}")
                    print(f"Translated: {translated_line}\n")
                    opfile.write(f"Original: {line}\n Translated: {translated_line}\n\n")
                else:
                    opfile.write('\n')  # Maintain blank lines

# Example usage
file_path = text_file_path
output_file = 'translated_text_file.txt'

translate_file_line_by_line(file_path, model, tokenizer)

print(f'Translation complete. Check the file {output_file} for the translated text.')

```

# Sentence by sentence translate English to Hindi

```python
from transformers import MarianMTModel, MarianTokenizer

# Load pre-trained model and tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-hi'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_text(text, model, tokenizer):
    # Tokenize and translate
    tokens = tokenizer(text, return_tensors="pt", padding=True)
    translated = model.generate(**tokens)
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text

def translate_file_sentence_by_sentence(file_path, model, tokenizer):
    with open(file_path, 'r', encoding='utf-8') as ipfile, open(output_file, 'a', encoding='utf-8') as opfile:
        for line in ipfile:
            line = line.strip()
            if line:
                sentences = line.split('.;? ')
                for sentence in sentences:
                    sentence = sentence.strip()
                    if sentence:
                        if not sentence.endswith('.;?'):
                            sentence += ' '
                        translated_sentence = translate_text(sentence, model, tokenizer)
                        print(f"Original: {sentence}")
                        print(f"Translated: {translated_sentence}\n")
                        opfile.write(f"Original: {sentence}\nTranslated: {translated_sentence}\n\n")
            else:
                opfile.write('\n')  # Maintain blank lines

# Example usage
file_path = '/content/4_5850433412128576047.txt'
output_file = 'translated_hindi_text_file.txt'

translate_file_sentence_by_sentence(file_path, model, tokenizer)

print(f'Translation complete. Check the file {output_file} for the translated text.')
```
